```python
from collections import Counter
import string
from datasets import load_dataset

dataset = load_dataset("wikitext", "wikitext-2-raw-v1")
text = dataset['train']['text']

def count_char_frequency(texts):
    counter = Counter()
    # Add digits and common symbols to the set of characters to count
    valid_chars = string.ascii_lowercase + string.ascii_uppercase + string.digits + string.punctuation
    for text in texts:
        counter.update(c for c in text if c in valid_chars)
    return counter

# Flatten the dataset into a list of text entries
all_texts = dataset['train']['text']  # Use the 'train' split

# Count letter frequencies
letter_counts = count_char_frequency(all_texts)

# Calculate the total number of characters
total_characters = sum(letter_counts.values())

# Sort by frequency (descending)
sorted_counts = sorted(letter_counts.items(), key=lambda item: item[1], reverse=True)

# Display results with percentages
print(f"Total characters: {total_characters}")
for letter, count in sorted_counts:
    percentage = (count / total_characters) * 100  # Calculate percentage
    print(f"{letter}: {count} ({percentage:.2f}%)")
```
